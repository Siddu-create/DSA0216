```python
# Step 1: Import Libraries
from google.colab import files
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Step 2: Upload Image
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Step 3: Read Image
img = cv2.imread(filename)

#  Resize Image (IMPORTANT  prevents very large display)
h, w = img.shape[:2]
scale_percent = 50   # Reduce to 50% of original size

new_width = int(w * scale_percent / 100)
new_height = int(h * scale_percent / 100)

img = cv2.resize(img, (new_width, new_height))

# Convert to grayscale (required for Canny)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# -----------------------------
# Step 4: Generate Gaussian Pyramid
# -----------------------------

level0 = gray
level1 = cv2.pyrDown(level0)
level2 = cv2.pyrDown(level1)
level3 = cv2.pyrDown(level2)

# -----------------------------
# Step 5: Apply Canny Edge Detection
# -----------------------------

edges0 = cv2.Canny(level0, 100, 200)
edges1 = cv2.Canny(level1, 100, 200)
edges2 = cv2.Canny(level2, 100, 200)
edges3 = cv2.Canny(level3, 100, 200)

# -----------------------------
# Step 6: Display Results
# -----------------------------

print("Level 0 - Original")
cv2_imshow(level0)
print("Edges Level 0")
cv2_imshow(edges0)

print("Level 1 - Half Size")
cv2_imshow(level1)
print("Edges Level 1")
cv2_imshow(edges1)

print("Level 2 - Quarter Size")
cv2_imshow(level2)
print("Edges Level 2")
cv2_imshow(edges2)

print("Level 3 - One-Eighth Size")
cv2_imshow(level3)
print("Edges Level 3")
cv2_imshow(edges3)
```
